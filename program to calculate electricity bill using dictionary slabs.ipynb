```python
# Function to calculate electricity bill using dictionary slabs
def electricity_bill(units):
    bill = 0

    # Slab rates stored in dictionary
    slabs = {
        100: 2,
        200: 3,
        float('inf'): 5
    }

    prev_limit = 0

    for limit, rate in slabs.items():
        if units > limit:
            bill += (limit - prev_limit) * rate
            prev_limit = limit
        else:
            bill += (units - prev_limit) * rate
            break

    return bill

# Input
u = int(input("Enter units: "))

# Output
print("Total bill amount =", electricity_bill(u))
```
