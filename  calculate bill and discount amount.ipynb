```python
# Function to calculate bill and discount amount
def shop_bill():
    price = float(input("Enter price: "))
    discount = float(input("Enter discount (%): "))

    discount_amt = (price * discount) / 100
    final_price = price - discount_amt

    return discount_amt, final_price

# Function call
d_amt, f_price = shop_bill()

# Output
print("Discount Amount =", d_amt)
print("Final price after discount =", f_price)
```
