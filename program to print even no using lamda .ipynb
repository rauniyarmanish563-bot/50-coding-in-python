```python
# Function to print even numbers using filter + lambda
def even_numbers(n):
    nums = range(1, n + 1)
    evens = list(filter(lambda x: x % 2 == 0, nums))
    print("Even numbers:", evens)

# Input
num = int(input("Enter N: "))

# Function call
even_numbers(num)
```
