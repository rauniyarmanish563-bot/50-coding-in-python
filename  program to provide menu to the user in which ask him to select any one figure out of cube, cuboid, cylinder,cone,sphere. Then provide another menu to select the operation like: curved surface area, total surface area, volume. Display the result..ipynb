```python
import numericalculation as nc

while True:
    print("\nSelect Figure:")
    print("1. Cube")
    print("2. Cuboid")
    print("3. Cylinder")
    print("4. Cone")
    print("5. Sphere")
    print("6. Exit")

    choice = int(input("Enter choice: "))

    if choice == 6:
        break

    print("\nSelect Operation:")
    print("1. Curved Surface Area")
    print("2. Total Surface Area")
    print("3. Volume")

    op = int(input("Enter operation: "))

    if choice == 1:
        a = float(input("Enter side: "))
        if op == 1:
            print("CSA:", nc.cube_csa(a))
        elif op == 2:
            print("TSA:", nc.cube_tsa(a))
        elif op == 3:
            print("Volume:", nc.cube_volume(a))

    elif choice == 2:
        l = float(input("Enter length: "))
        b = float(input("Enter breadth: "))
        h = float(input("Enter height: "))
        if op == 1:
            print("CSA:", nc.cuboid_csa(l, b, h))
        elif op == 2:
            print("TSA:", nc.cuboid_tsa(l, b, h))
        elif op == 3:
            print("Volume:", nc.cuboid_volume(l, b, h))

    elif choice == 3:
        r = float(input("Enter radius: "))
        h = float(input("Enter height: "))
        if op == 1:
            print("CSA:", nc.cylinder_csa(r, h))
        elif op == 2:
            print("TSA:", nc.cylinder_tsa(r, h))
        elif op == 3:
            print("Volume:", nc.cylinder_volume(r, h))

    elif choice == 4:
        r = float(input("Enter radius: "))
        l = float(input("Enter slant height: "))
        h = float(input("Enter height: "))
        if op == 1:
            print("CSA:", nc.cone_csa(r, l))
        elif op == 2:
            print("TSA:", nc.cone_tsa(r, l))
        elif op == 3:
            print("Volume:", nc.cone_volume(r, h))

    elif choice == 5:
        r = float(input("Enter radius: "))
        if op == 1:
            print("CSA:", nc.sphere_csa(r))
        elif op == 2:
            print("TSA:", nc.sphere_tsa(r))
        elif op == 3:
            print("Volume:", nc.sphere_volume(r))

    else:
        print("Invalid choice")
```
