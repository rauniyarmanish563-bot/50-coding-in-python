```python
# Function to check voting eligibility with validation
def check_vote():
    while True:
        try:
            age = int(input("Enter age: "))

            if age < 0:
                print("Invalid age, try again!")
                continue

            if age >= 18:
                print("Eligible")
            else:
                print("Not Eligible")

            break   # exit loop after valid input

        except ValueError:
            print("Please enter a valid number!")

# Function call
check_vote()
```
