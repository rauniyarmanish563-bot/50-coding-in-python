```python
# Function to print even numbers using while loop
def print_even(n):
    i = 2   # start from first even number

    while i <= n:
        print(i, end=" ")
        i += 2   # jump to next even number

# Input
num = int(input("Enter N: "))

# Function call
print_even(num)
```
