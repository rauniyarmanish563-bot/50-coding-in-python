```python
# Function using lambda for simple interest
def simple_interest():
    # Taking input
    p = float(input("Enter P: "))
    r = float(input("Enter R: "))
    t = float(input("Enter T: "))

    # Lambda function
    si = lambda p, r, t: (p * r * t) / 100

    print("Simple Interest =", si(p, r, t))

# Function call
simple_interest()
```
