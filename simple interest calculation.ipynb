```python
# Function to calculate simple interest (return-based)
def calculate_si(p, r, t):
    return (p * r * t) / 100

# Input
P = float(input("Enter Principal: "))
R = float(input("Enter Rate: "))
T = float(input("Enter Time: "))

# Function call
SI = calculate_si(P, R, T)

# Output
print("Simple Interest =", SI)
```
