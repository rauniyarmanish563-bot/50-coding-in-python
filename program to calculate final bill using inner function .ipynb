```python
# Function to calculate final bill using inner function
def shopkeeper_bill():

    # Inner function for calculation
    def calculate(price, discount):
        return price - (price * discount / 100)

    # Input
    price = float(input("Enter price: "))
    discount = float(input("Enter discount (%): "))

    # Validation
    if discount < 0 or discount > 100:
        print("Invalid discount!")
        return

    # Output
    print("Final price after discount =", calculate(price, discount))

# Function call
shopkeeper_bill()
```
