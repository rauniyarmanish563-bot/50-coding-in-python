```python
# Function to calculate final bill after discount
def calc_final(price, discount):
    # Using direct formula
    return price - (price * discount / 100)

# Input
p = float(input("Enter price: "))
d = float(input("Enter discount (%): "))

# Output using formatted string
print(f"Final price after discount = {calc_final(p, d)}")
```
