```python
# Function to get even numbers using list comprehension
def get_even(n):
    even_list = [i for i in range(1, n+1) if i % 2 == 0]
    print("Even numbers:", even_list)

# Input
num = int(input("Enter N: "))

# Function call
get_even(num)
```
