```python
# Function to calculate electricity bill
def calc_bill(units):
    bill = 0

    for i in range(1, units + 1):
        if i <= 100:
            bill += 2
        elif i <= 200:
            bill += 3
        else:
            bill += 5

    print("Total bill amount =", bill)

# Input
u = int(input("Enter units: "))

# Function call
calc_bill(u)
```
