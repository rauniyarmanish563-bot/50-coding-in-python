```python
# Function to calculate electricity bill using match-case
def electricity_bill(units):
    match units:
        case u if u <= 100:
            bill = u * 2
        case u if u <= 200:
            bill = (100 * 2) + (u - 100) * 3
        case _:
            bill = (100 * 2) + (100 * 3) + (units - 200) * 5

    return bill

# Input
u = int(input("Enter units: "))

# Output
print("Total bill amount =", electricity_bill(u))
```
