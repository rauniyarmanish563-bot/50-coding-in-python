```python
# Function to print even numbers from 1 to N
def print_even_numbers(n):
    print("Even numbers between 1 and", n, "are:")

    # Loop from 1 to n
    for i in range(1, n + 1):
        # Check if number is even
        if i % 2 == 0:
            print(i, end=" ")

# Taking input from user
N = int(input("Enter N: "))

# Calling the function
print_even_numbers(N)
```

```python
# Function to check eligibility
def check_vote_eligibility(age):
    if age >= 18:
        return "Eligible"
    else:
        return "Not Eligible"

# Taking input from user
age = int(input("Enter age: "))

# Calling function and printing result
result = check_vote_eligibility(age)
print(result)
```

```python
# Function to check voting eligibility (different method)
def check_vote(age):
    # Using ternary operator
    result = "Eligible to vote" if age >= 18 else "Not Eligible to vote"

    # Print inside function
    print(result)

# Taking input
age = int(input("Enter age: "))

# Function call
check_vote(age)
```

```python
# Function to calculate electricity bill
def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 1.5
    elif units <= 200:
        bill = (100 * 1.5) + (units - 100) * 2.5
    else:
        bill = (100 * 1.5) + (100 * 2.5) + (units - 200) * 4

    return bill

# Taking input from user
units = int(input("Enter units consumed: "))

# Calling function
total_bill = calculate_bill(units)

# Display result
print("Total electricity bill is: ₹", total_bill)
```

```python
# Function to calculate simple interest
def simple_interest(p, r, t):
    si = (p * r * t) / 100
    print("Simple Interest is:", si)

# Taking input from user
p = float(input("Enter Principal (P): "))
r = float(input("Enter Rate (R): "))
t = float(input("Enter Time (T): "))

# Function call
simple_interest(p, r, t)
```
