```python
# Function to calculate simple interest using map
def simple_interest(p, r, t):
    values = list(map(float, [p, r, t]))
    si = (values[0] * values[1] * values[2]) / 100
    return si

# Input
p = input("Enter P: ")
r = input("Enter R: ")
t = input("Enter T: ")

# Output
print("Simple Interest =", simple_interest(p, r, t))
```
