```python
# Function to calculate electricity bill
def bill_amount(units):
    if units <= 100:
        return units * 2
    elif units <= 200:
        return (100 * 2) + (units - 100) * 3
    else:
        return (100 * 2) + (100 * 3) + (units - 200) * 5

# Input
u = int(input("Enter units: "))

# Function call + output in one line
print("Total bill amount =", bill_amount(u))
```
