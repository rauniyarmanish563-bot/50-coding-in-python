```python
# Generator function to yield even numbers
def even_gen(n):
    for i in range(2, n + 1, 2):
        yield i

# Input
num = int(input("Enter N: "))

# Using generator
print("Even numbers:", list(even_gen(num)))
```

```python
# Class to calculate simple interest
class Interest:

    def __init__(self, p, r, t):
        self.p = p
        self.r = r
        self.t = t

    def calculate(self):
        return (self.p * self.r * self.t) / 100

# Input
p = float(input("Enter Principal: "))
r = float(input("Enter Rate: "))
t = float(input("Enter Time: "))

# Object creation
obj = Interest(p, r, t)

# Output
print("Simple Interest =", obj.calculate())
```

```python
# Recursive function to calculate final price
def final_price(price, discount):
    if discount == 0:
        return price
    return final_price(price - (price / 100), discount - 1)

# Input
p = float(input("Enter price: "))
d = int(input("Enter discount (%): "))

# Output
print("Final price after discount =", final_price(p, d))
```
