```python
# Function to get even numbers using step in range
def even_numbers(n):
    print("Even numbers are:")

    # Start from 2 and jump by 2
    for i in range(2, n + 1, 2):
        print(i, end=" ")

# Input
num = int(input("Enter N: "))

# Function call
even_numbers(num)
```
