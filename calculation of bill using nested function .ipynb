```python
# Function to calculate bill using nested function
def electricity_bill(units):

    def calculate(u):
        if u <= 100:
            return u * 1.8
        elif u <= 200:
            return (100 * 1.8) + (u - 100) * 2.8
        else:
            return (100 * 1.8) + (100 * 2.8) + (u - 200) * 4.5

    return calculate(units)

# Input
units = int(input("Enter units: "))

# Output
print("Total bill amount =", electricity_bill(units))
```
