```python
# Function using lambda check eligibility for voting
def check_vote(age):
    result = (lambda x: "Eligible" if x >= 18 else "Not Eligible")(age)
    print(result)

# take Input from user
age = int(input("Enter age: "))

# Function call
check_vote(age)
```
