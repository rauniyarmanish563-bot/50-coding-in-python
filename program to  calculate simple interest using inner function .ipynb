```python
# Function to calculate simple interest using inner function
def simple_interest():

    def calculate(p, r, t):
        return (p * r * t) / 100

    # Input
    p = float(input("Enter P: "))
    r = float(input("Enter R: "))
    t = float(input("Enter T: "))

    # Output
    print("Simple Interest = {:.2f}".format(calculate(p, r, t)))

# Function call
simple_interest()
```
